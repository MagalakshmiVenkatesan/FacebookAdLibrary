```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

# Fianna Fail

```python
#import FiannaFail dataset

df = pd.read_excel('D:/NUIG/Final Project/FinalData/FiannaFail_Final.xlsx')

df.describe()
```

```python
#FiannFail box plot 

bp = sns.boxplot(x='Problem Sector',y='AvgImpressions',data=df)

bp.set_xlabel("Problem Sector",fontsize=20)

bp.set_ylabel("AvgImpressions",fontsize=20)

plt.xticks(rotation=90)

plt.ylim([-500,250000])
```

```python
# Fianna Fail correlation matrix

correlation_matrix = df.corr()
fig, ax = plt.subplots(figsize=(10,10)) 
sns.heatmap(correlation_matrix, annot=True, linewidths=.5, ax=ax)
plt.show()
```

```python
# Fianna Fail Regression plot

rs= sns.regplot(x='Votes',y='AvgImpressions',data=df)
plt.xlim([450000,700000])
rs.set_xlabel("Problem Sector",fontsize=20)

rs.set_ylabel("AvgImpressions",fontsize=20)
```

```python
rs= sns.regplot(x='Votes',y='AvgAmount',data=df)
plt.xlim([450000,700000])
rs.set_xlabel("Problem Sector",fontsize=20)

rs.set_ylabel("AvgAmount",fontsize=20)
```
