```python
bp = sns.boxplot(x='Problem Sector',y='AvgImpressions',data=df)

bp.set_xlabel("Problem Sector",fontsize=20)

bp.set_ylabel("AvgImpressions",fontsize=20)

plt.xticks(rotation=90)

plt.ylim([-500,250000])
```
